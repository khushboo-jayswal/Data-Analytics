```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.boxplot(y=df['YearlyIncome'])       # boxenplot
plt.title("Boxplot of Yearly Income")
plt.show()
```

```python
sns.violinplot(df['YearlyIncome'])       # boxenplot
plt.title("Violinplot of Yearly Income")
plt.show()
```

```python
sns.distplot(df['YearlyIncome'])
plt.show()
```

```python
sns.distplot(df['YearlyIncome'], hist=False)
plt.show()
```

```python
sns.distplot(df['YearlyIncome'], bins=10, color='g', hist_kws=dict(edgecolor='r',linewidth=3))
plt.show()
```

```python
sns.countplot(df['BikeBuyer'])
plt.show()
```

```python
sns.countplot(df['NumberCarsOwned'], palette=sns.cubehelix_palette())
plt.show()
```
